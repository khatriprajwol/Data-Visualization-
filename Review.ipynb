```python
%%bigquery --project data-base-1-338717
SELECT *
FROM `bigquery-public-data.london_fire_brigade.fire_brigade_service_calls`
LIMIT 15
```

This input gives me only 15 rows even though it has total of 32247 rows. That is really helpful to over view in a small table which beautifully presents the information. 

```python
%%bigquery --project data-base-1-338717 
SELECT Count(*) as number_of_late_night_false_alarm
FROM `bigquery-public-data.london_fire_brigade.fire_brigade_service_calls`
WHERE incident_group = "False Alarm" AND (hour_of_call >= 22 OR hour_of_call <= 5)
```

```python
%%bigquery --project data-base-1-338717 
SELECT incident_group, AVG(num_pumps_attending) as avg_pumps_attending
FROM `bigquery-public-data.london_fire_brigade.fire_brigade_service_calls`
GROUP BY incident_group
```

```python
%%bigquery --project data-base-1-338717
SELECT property_category, MAX(num_pumps_attending) as max_pumps_attending
FROM `bigquery-public-data.london_fire_brigade.fire_brigade_service_calls`
GROUP BY property_category
```
