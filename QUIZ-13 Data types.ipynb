```python
import pandas as pa

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/bank.csv')

df.head()
```

```python
df.info()
```

The first column which is `age` is an integer. I really do not think the change is necessary. If we want then we can create the correlation between `age` with other attributes. 

```python
from pandas.api.types import CategoricalDtype

cat_type = CategoricalDtype(categories=['primary', 'secondary', "tertiary"], ordered=True)

dfeducation = df.education.astype(cat_type)

dfeducation.dtype
```

`Job` is a `categorical` varibale we can not give it an order because even if we do it won't make sense. However, `Education` is a `categorical` variable. We can certainly give it an order. The way I have done is shown in the above table.
