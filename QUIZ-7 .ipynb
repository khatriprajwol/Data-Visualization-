Here is the austin data. 

```python
%%bigquery --project data-base-1-338717
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
LIMIT 10 
```

```python
%%bigquery --project data-base-1-338717

SELECT bikeid, AVG(duration_minutes) as average, STDDEV(duration_minutes) as standarddeviation 
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY bikeid
```

```python
%%bigquery --project data-base-1-338717

SELECT MAX(duration_minutes) as max_time
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
```

This a london bicycle data. 

```python
%%bigquery --project data-base-1-338717
SELECT *
FROM `bigquery-public-data.london_bicycles.cycle_hire`
LIMIT 10 
```

```python
%%bigquery --project data-base-1-338717 

SELECT bike_id, AVG(duration) as average, STDDEV(duration) as standarddeviation 
FROM `bigquery-public-data.london_bicycles.cycle_hire`
GROUP BY bike_id
LIMIT 10
```

```python
%%bigquery --project data-base-1-338717

SELECT MAX(duration) as max_time2
FROM `bigquery-public-data.london_bicycles.cycle_hire`
```
