These are the libraries I have used to import the HTML code. 

```python
import requests
import pandas as pa
from bs4 import BeautifulSoup
```

The code below helps me to request the information of html code. 

```python
r = requests.get('https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_films')
html_contents = r.text
html_soup = BeautifulSoup(html_contents,"lxml")
```

```python
len(html_soup.find_all('table'))
```

As we can see there are only seven tables. 

```python
tables = html_soup.find_all('table',class_="wikitable plainrowheaders")
tables[0].find_all('a')
```

These are the list of urls in the html. 

```python
df = pa.read_html(str(tables))[0]
df
```

I will use the droplevel command twice to remove two of marbel movies form the table.

```python
df.columns = df.columns.droplevel(1)
df
```

Here I have removed 6 and 13 from the table and below is the code used to do it. 

```python
df = df.drop(6).drop(13)
df
```
