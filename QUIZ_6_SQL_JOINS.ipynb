However, I have used JOIN to combine two data for parkland and sidewalk.

```python
%%bigquery --project data-base-1-338717
WITH t as(
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_stations`
WHERE property_type= 'parkland')

SELECT *
FROM t as table1  JOIN `bigquery-public-data.austin_bikeshare.bikeshare_trips` as table2 ON table1.name= table2.start_station_name
limit 20
```

```python
%%bigquery --project data-base-1-338717
WITH t as(
SELECT * 
FROM `bigquery-public-data.austin_bikeshare.bikeshare_stations`)


SELECT *
FROM t as table1 LEFT OUTER JOIN `bigquery-public-data.austin_bikeshare.bikeshare_trips` as table2 ON table1.name= table2.start_station_name
WHERE table1.property_type='sidewalk' OR table1.property_type='parkland'
limit 100
```

```python
%%bigquery --project data-base-1-338717
WITH t as(
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_stations`)

SELECT *
FROM t as table1 RIGHT OUTER JOIN `bigquery-public-data.austin_bikeshare.bikeshare_trips` as table2 ON table1.name= table2.start_station_name
limit 100
```
