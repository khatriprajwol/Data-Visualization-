```python
import pandas as pa

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')
```

```python
df.calories.plot.box()
```

```python
df.boxplot(column = 'calories',by = 'workout_type',rot = 45)
```

```python
df.loc[df["workout_type"]== "Trekking"]
```

```python
by_Range = df.groupby("workout_type")
by_Range.get_group("Trekking")
```

```python
by_Range.boxplot(subplots='Trekking', column='calories', fontsize=10, rot=35, grid=True, ax=None, figsize= '10,10', layout=None, sharex=False, sharey=True, backend=None) 
```

```python
df.head()
```

```python
df.boxplot(column = 'calories',by = 'workout_type',vert = False)
```

```python
by_Range.boxplot(column = 'calories',by = 'workout_type',vert = True, rot=None, figsize=(10,10), sharex=True, sharey=True, backend=('Trekking'))
```

```python
import seaborn as sns

ax = sns.boxplot(data = df, x = 'workout_type', y = 'calories')
ax.set_xticklabels(ax.get_xticklabels(),rotation = 45)
```

```python
sns.boxplot( y = df.workout_type, x = df.calories)
```
