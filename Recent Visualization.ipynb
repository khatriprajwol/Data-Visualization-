```python
import requests
import pandas as pa
from bs4 import BeautifulSoup
r = requests.get("https://en.wikipedia.org/wiki/Forbes%27_list_of_the_most_valuable_sports_teams")
html_contents = r.text
html_soup = BeautifulSoup(html_contents,"lxml")
```

```python
tables = html_soup.find_all('table')
df = pa.read_html(str(tables))[5]
df.head(10)
```

##**What is excellence in term of statistics?**
[Excellence](https://github.com/nurfnick/Data_Viz/blob/main/21_Graphical_Excellence.ipynb) in statistical graphics consists of complex ideas communicated with clarity, precision and efficiency". Excellence of graph enhances elegance. According to the Edward Tufte's books on the [Visual Display of Quantitative Information](http://www.econ.upf.edu/~michael/visualdata/tufte-aesthetics_and_technique.pdf) (Graphics Press 2001), "Graphical elegance is often found in simplicity of design and complexity of data."


##**Graphical Excellence of Both Graph** 
Graphical excellence is crucial and a great technique to make a critique about the graph. It also gives and idea about what a graph should contain and helps us determine if it is a good graph. In my project in order to generalize if above graphs are excellent, I have used the idea from the Edward Tufte's books on the [Visual Display of Quantitative Information](http://www.econ.upf.edu/~michael/visualdata/tufte-aesthetics_and_technique.pdf) (Graphics Press 2001) where he mentions the road toward best graphical excellence. According to Tufte they are:
* well-designed presentation of data of substance, statistics and design.
* complex ideas communicated with clarity, precision and efficiency
* the greatest number of ideas in the shortest time with the least ink in the smallest space.
